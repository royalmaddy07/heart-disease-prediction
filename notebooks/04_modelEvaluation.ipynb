Import Libraries

```python
import joblib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# NOTE: for each classification model used in this project -> I have evaluated there train-test accuracy, classification
#       report and confusion matrix
# train-test accuracy -> will tell whether the model is overfitting, underfitting or generalizing
# classification report -> will give us the performance of the model based on each class
# confusion matrix -> will show which classes are confused with others
```

Model Evaluation

```python
# loading models from saved .pkl files ->
lr_model = joblib.load("../models/lr_model.pkl")
knn_model = joblib.load("../models/knn_model.pkl")
dt_model = joblib.load("../models/dt_model.pkl")
rf_model = joblib.load("../models/rf_model.pkl")
gb_model = joblib.load("../models/gb_model.pkl")
xgb_model = joblib.load("../models/xgb_model.pkl")
nb_model = joblib.load("../models/nb_model.pkl")

# loading testing data from stored processed files ->
X_train = pd.read_csv("../data/processed/X_train.csv")
y_train = pd.read_csv("../data/processed/y_train.csv")
X_test = pd.read_csv("../data/processed/X_test.csv")
y_test = pd.read_csv("../data/processed/y_test.csv")
```

Logistic Regression Classifier

```python
y_pred_lr = lr_model.predict(X_test)

accuracy_lr = accuracy_score(y_test,y_pred_lr)

print(f"train accuracy : {lr_model.score(X_train,y_train)*100:.4f}%")
print(f"test accuracy : {accuracy_lr*100:.4f}% \n") 

print(classification_report(y_test,y_pred_lr))

cm = confusion_matrix(y_test,y_pred_lr)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1,2,3,4])
plt.figure(figsize=(6,6))
disp.plot(cmap='Purples', values_format='d')
plt.title('Confusion Matrix for Logistic Regression Model')
plt.show()
```

```python
y_pred_knn = knn_model.predict(X_test)

accuracy_knn = accuracy_score(y_test,y_pred_knn)
print(f"train accuracy : {knn_model.score(X_train,y_train)*100:.4f}%")
print(f"test accuracy : {accuracy_knn*100:.4f}%")

print(classification_report(y_test,y_pred_knn))

cm = confusion_matrix(y_test,y_pred_knn)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[0,1,2,3,4])
plt.figure(figsize=(6,6))
disp.plot(cmap='Purples', values_format='d')
plt.title('Confusion Matrix for k-nearest neighbours Model')
plt.show()
```

Model Comparision

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score

models = {
    'Logistic' : lr_model,
    'KNN' : knn_model,
    'Decision_Trees' : dt_model,
    'random_Forest' : rf_model,
    'Gradient_boosting' : gb_model,
    'XGBoost' : xgb_model,
    'Naive_Bayes' : nb_model
}

names = []
trainAccuracy = []
testAccuracy = []
recall = []
f1Score = []
precision = []

for name,model in models.items():
    y_pred = model.predict(X_test)# predicting the values

    names.append(name)# adding the name

    trainAccuracy.append(round(model.score(X_train,y_train)*100, 2))# train accuracy

    testAccuracy.append(round(accuracy_score(y_test,y_pred)*100, 2))# test accuracy

    recall.append(round(recall_score(y_test,y_pred,average='macro')*100, 2))# 

    f1Score.append(round(f1_score(y_test,y_pred,average='macro')*100, 2))

    precision.append(round(precision_score(y_test,y_pred,average='macro')*100, 2))

df = pd.DataFrame({
    'Model' : names,
    'Train Accuracy (%)' : trainAccuracy,
    'Test Accuracy (%)' : testAccuracy,
    'Macro Recall (%)' : recall,
    'Macro Precision (%)' : precision,
    'Macro F1 Score (%)' : f1Score
})
```

```python
display(df)

df_sorted_recall = df.sort_values(by=['Macro Recall (%)'],ascending=False).set_index('Model')

display(df_sorted_recall)

df_sorted_accuracy = df.sort_values(by=['Test Accuracy (%)'],ascending=False).set_index('Model')

display(df_sorted_accuracy)
```
